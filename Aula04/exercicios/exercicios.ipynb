1. Crie um programa que possa incluir, pesquisar, alterar e excluir nomes em uma lista, que pode ter quantos nomes o usuário desejar.


```python
#importa a blibioteca
from time import sleep as delay
#var
lst_name = []
opcao : int
print(f"{"-"*25} Gerenciador de Pessoas {"-"*25}")

#permite o user "criar" a lista
while True:
    nome_inp = input("Digite os nomes que deseja quando terminar aperda espaço sem nada ná frente: ")
    if nome_inp == "":
        break
    lst_name.append(nome_inp)

#menu de opções para ver o que o usuário vai fazer
while True:
    #deixa a lista ordenada
    lst_name.sort()
    delay(1.5)
    #limpa o terminal pela sequência ANSI
    print('\033[H\033[J', end="")
    #imprime os nomes da lista caso a lista tenha elementos
    print(f"Lista atual: {" ,".join(lst_name)}." if lst_name != [] else "Sem nomes na lista.")
    print("1. Adicionar nomes na lista")
    print("2. Pesquisar por nome na lista")
    print("3. Alterar nome na lista")
    print("4. Excluir nome da lista")
    print("5. Sair do programa")
    #tratamento de erro
    try:
        opcao = int(input("O que você deseja fazer agora? "))
    except ValueError:
        print("Isso não é um número!")
        opcao = -1
    #faz o que o usuário quer 
    if opcao == 1:
        print('\033[H\033[J', end="")
        print(f"Lista atual: {",".join(lst_name)}")
        #usuário adiciona nomes na lista
        while True:
            nome_inp = input("Digite os nomes desejados. Para terminar, pressione ENTER sem digitar nada: ")
            if nome_inp == "":
                break
            lst_name.append(nome_inp)
    elif opcao == 2: 
        print('\033[H\033[J', end="")
        print(f"Lista atual: {",".join(lst_name)}")
        nome = input("Nome que deseja pesquisar: ")
        #ver se o nome existe na lista
        if nome in lst_name:
            print(f"'{nome}' já existe na lista")
        else:
            print("Nome não existe na lista")
    elif opcao == 3: 
        print('\033[H\033[J', end="")
        print(f"Lista atual: {",".join(lst_name)}")
        nome = input("Qual é o nome que você deseja alterar na lista? ")
        novo_nome = input("Novo nome: ")
        if nome in lst_name:
            #pega o index 
            index = lst_name.index(nome)
            #é coloca ele
            lst_name[index] = novo_nome
            print(f"O nome '{nome}' foi alterado para '{novo_nome}' com sucesso.")
        else:
            print("Nome não está na lista, então não pode ser alterado.\nA lista não foi modificada.")
    elif opcao == 4: 
        print('\033[H\033[J', end="")
        print(f"Lista atual: {",".join(lst_name)}")
        user_inp = input("Qual é o nome que você deseja remover da lista? ")
        if user_inp in lst_name:
            lst_name.remove(user_inp)
            print(f"O nome '{user_inp}' foi removido com sucesso.")
        else:
            print("Nome não está na lista, então não pode ser removido.\nA lista não foi modificada.")
    elif opcao == 5: 
        print('\033[H\033[J', end="")
        print("Tchau!")
        print(f"Lista final: {",".join(lst_name)}")
        #encerra o programa
        exit()
    #isso é só para o user voltar e digitar um número 
    elif opcao == -1: continue
    else:
        print("Opção inválida!")
```

2. Crie um programa que o usuário possa digitar qauntos números digitar, e ao terminar, imprima os números em ordem crescente.


```python
#var
lst_n = []
print(20*"-","Ordenação de 4 números",20*"-")
while True:
    i = input("Digte os número; quando termina der um espaço vazio: ")
    '''Por que não usar o try except?
        orque o .join() só funciona quando a variável é uma string'''
    while not i.isdigit():
        i = input(f"Digite um número válido.: ")
    lst_n.append(i)
    if i == "":
        break

#mostra para o usuário a lista formatada e ordenada
print(f"Lista ordenada: {", ".join(sorted(lst_n))}")
```

3. Crie um programa que o usuário possa digitar uma quantidade desejada de notas de um determinado aluno (nota mínima 0 e nota máxima 10), e o programa calcula a média desse aluno, e ao final informe se o aluno está aprovado ou reprovado (média mínima para aprovação = 7).

```python
#importações de bibliotcas
from time import sleep as delay
from pandas import DataFrame as df
#var      
tabela = {
    #coluna: #linha
    "Alunos":[],
    "1°Bimestre":[],
    "2°Bimestre":[],
    "3°Bimestre":[],
    "4°Bimestre":[],
    "Nota Final":[],
    "Situação":[]
}
print(f"{15*"-"} Tabela de notas. {15*"-"}\n")
while True:
    #inputs para o user:
    nome_aluno = input("Insira o nome do aluno: ")
    tabela["Alunos"].append(nome_aluno)
    #repete 4 vezes
    for i in range(1,5): 
        #entrada do usuário
        notas = input(f"Digite o {i}º número(Max: 10 e Min:0): ").replace(",",".")
        while True:
            try:
                notas = float(notas)
                break
            except ValueError:
                notas = input("Digite um número válido: ")
        #aqui eu forço o user digita um valor entre 0 á 10
        while not notas in range(11):
            notas = input("Digite um número válido: ")
            while True:
                try:
                    notas = float(notas)
                    break
                except ValueError:
                    notas = input("Digite um número válido: ")
        #add nós bimestre ná lista certa
        if i == 1:
            tabela["1°Bimestre"].append(notas)
        elif i == 2:
            tabela["2°Bimestre"].append(notas)
        elif i == 3:
            tabela["3°Bimestre"].append(notas)
        elif i == 4:
            tabela["4°Bimestre"].append(notas)
    #pega na tabela o utima indece para fazer a média
    media = (tabela["1°Bimestre"][-1]+tabela["2°Bimestre"][-1]+tabela["3°Bimestre"][-1]+tabela["4°Bimestre"][-1])/4
    tabela["Nota Final"].append(media)
    #ver a situação do aluno
    if media < 7:
        tabela["Situação"].append("Reprovado")
    else:
        tabela["Situação"].append("Aprovado")
    '''organisa(alfabetica/cresente) a linha com base na coluna
    o by= fala qual é a coluna'''
    tabela_df = df(tabela).sort_values(by="Alunos")
    '''converte o dic numa espese de "panilhia"(dataframe);
    .to_sting() --> server para transforma esse 
    formado de 'panilhia' em str;
    o idex=False --> tira o idex antes do nome;
    o justify="center" --> centraliza com base na coluna;
    '''
    print(tabela_df.to_string(index=False,justify="center").replace(".",","))
    #ver se o user deseja finaliza
    ipunt_user = input("\n deseja digitar mas alunos(s-sim|n-não)? ").lower()
    if ipunt_user != "s":
        break
    delay(1.5)
    #limpa o terminal pela sequência ANSI
    print('\033[H\033[J', end="")
    
print('\033[H\033[J', end="")
print("Nota Finais: ",tabela_df.to_string(index=False,justify="center").replace(".",","), sep=2*"\n")
```
